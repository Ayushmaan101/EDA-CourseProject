### Load Libraries and Dataset

```python
# Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Loading the dataset from the provided URL
url = 'https://raw.githubusercontent.com/salemprakash/EDA/main/Data/student-mat.csv'
df = pd.read_csv(url)

# Display the first 5 rows
df.head()
```

### Initial Data Exploration

Dimensions of the dataset

```python
#Dimensions of the Dataset
print(f"The dataset has {df.shape[0]} rows and {df.shape[1]} columns.")
```

Summary Statistics

```python
print("Summary statistics for numerical columns:")
df.describe()
```

Data Types and Non-Null Counts

```python
print("Data types and non-null counts:")
df.info()
```

### Data Cleaning and Preprocessing

Handling Missing Data


```python
print("Missing values per column:")
print(df.isnull().sum())
```

Replacing Values

```python
df['romantic_numeric'] = df['romantic'].replace({'yes': 1, 'no': 0})
print("Value counts for the new 'romantic_numeric' column:")
print(df['romantic_numeric'].value_counts())
```

### Data Transformation and Reshaping

Renaming Axis Indexes (Columns)

```python
df_renamed = df.rename(columns={'G3': 'Final_Grade', 'Medu': 'Mother_Education'})
print("First 5 columns after renaming:", df_renamed.columns[:5])
```

Data Normalization

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler

# Min-Max Scaling (scales to a [0, 1] range)
min_max_scaler = MinMaxScaler()
df['absences_minmax'] = min_max_scaler.fit_transform(df[['absences']])

# Z-Score Standardization (scales to a mean of 0 and standard deviation of 1)
z_score_scaler = StandardScaler()
df['absences_zscore'] = z_score_scaler.fit_transform(df[['absences']])

print("Normalized 'absences' column head:\n", df[['absences', 'absences_minmax', 'absences_zscore']].head())
```

Discretization and Binning

```python
# Equal-Width Binning (each bin has the same age range)
df['age_binned_width'] = pd.cut(df['age'], bins=3, labels=['15-17', '18-19', '20-22'])
print("Equal-Width Binning Counts:\n", df['age_binned_width'].value_counts())

# Equal-Frequency Binning (each bin has roughly the same number of students)
df['age_binned_freq'] = pd.qcut(df['age'], q=4, labels=['Q1', 'Q2', 'Q3', 'Q4'])
print("\nEqual-Frequency Binning Counts:\n", df['age_binned_freq'].value_counts())
```

Outlier Detection and Filtering

```python
# Calculate IQR for 'absences'
Q1 = df['absences'].quantile(0.25)
Q3 = df['absences'].quantile(0.75)
IQR = Q3 - Q1

# Define outlier boundaries
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers = df[(df['absences'] < lower_bound) | (df['absences'] > upper_bound)]
print(f"Identified {len(outliers)} outliers in the 'absences' column based on the IQR method.")

# For analysis, you might create a new dataframe without these outliers
df_no_outliers = df[(df['absences'] >= lower_bound) & (df['absences'] <= upper_bound)]
print(f"Shape of dataframe after filtering outliers: {df_no_outliers.shape}")
```

Computing Indicator/Dummy Variables

```python
# Create dummy variables for the 'Mjob' (Mother's Job) column
dummies_mjob = pd.get_dummies(df['Mjob'], prefix='Mjob')
print("Dummy variables for Mjob:\n", dummies_mjob.head())

# We can join this back to the original dataframe if needed for modeling
df_with_dummies = pd.concat([df, dummies_mjob], axis=1)
```

Reshaping and Pivoting

```python
# Create a pivot table to see average final grade
pivot_table = df.pivot_table(values='G3', index='Mjob', columns='Fjob', aggfunc='mean')
print("Pivot Table: Average Final Grade by Parents' Jobs")
print(pivot_table)
```

### Permutation and Random Sampling

Permutation

```python
# Randomly shuffle the dataframe
df_shuffled = df.sample(frac=1).reset_index(drop=True)
print("Shuffled dataframe head:\n", df_shuffled.head())
```

Random Sampling

```python
# Random sampling WITHOUT replacement (each row can be chosen only once)
df_sample_no_replace = df.sample(n=5, replace=False)
print("Sample without replacement:\n", df_sample_no_replace)

# Random sampling WITH replacement (a row can be chosen multiple times)
df_sample_with_replace = df.sample(n=5, replace=True)
print("\nSample with replacement:\n", df_sample_with_replace)
```

### Univariate Analysis

Histogram for Final Grades

Box Plot for Absences

### Bivariate and Multivariate Analysis

Scatter Plot: Study Time vs. Final Grade

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='studytime', y='G3', data=df, hue='sex', style='internet')
plt.title('Study Time vs. Final Grade')
plt.xlabel('Weekly Study Time (Categorical)')
plt.ylabel('Final Grade (G3)')
plt.legend(title='Legend')
plt.show()
```

Correlation Heatmap

```python
# Select only numeric columns for the correlation matrix
numeric_cols = df.select_dtypes(include=np.number)

# Calculate Pearson correlation
pearson_corr = numeric_cols.corr(method='pearson')

# Plot the Pearson correlation heatmap
plt.figure(figsize=(20, 16))
sns.heatmap(pearson_corr, annot=True, cmap='coolwarm', fmt='.2f', linewidths=.5)
plt.title('Pearson Correlation Matrix of Numerical Variables')
plt.show()
```
